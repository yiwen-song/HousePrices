## Features selection

**'OverallQual', 'GrLivArea' and 'GarageCars' are highly correlateed with 'SalePrice', so they should be selected.  
'FullBath'is also correlated with 'SalePrice', so it should be selected.  
'GarageCars' and 'GarageArea' are correlated, so we only use 'GarageCars'.  
'YearBuilt' and 'YearRemoAdd', 'TotalBsmtSF' and '1stFlrSF' are correlated, so use the mean of each two variables.**

```python
train['YearBuiltRemod']=(train['YearBuilt']+train['YearRemodAdd'])/2
train['Bsmt1stSF']=(train['TotalBsmtSF']+train['1stFlrSF'])/2
cols= ['OverallQual', 'GrLivArea', 'GarageCars','FullBath', 'YearBuiltRemod','Bsmt1stSF']
```

```python
# plot pairwise relationship in features
sns.pairplot(train[cols])
```

# Modeling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
scaler=StandardScaler()
X=train[cols]
y=train['SalePrice']
scaled_X=scaler.fit_transform(X)
X_train,X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.25, random_state=42)
model=LinearRegression()
model.fit(X_train,y_train)
print(model.coef_)
print(model.intercept_)
```

```python
pred=model.predict(X_test)
plt.scatter(y_test, pred)
```

```python
# residual histogram
sns.distplot((y_test-pred), bins=50)
```

# Prediction

```python
test['YearBuiltRemod']=(test['YearBuilt']+test['YearRemodAdd'])/2
test['Bsmt1stSF']=(test['TotalBsmtSF']+test['1stFlrSF'])/2
test[cols].isna().sum()
```

```python
# fill null values with variable means
test['Bsmt1stSF'].fillna(test['Bsmt1stSF'].mean(),inplace=True)
test[cols].isna().sum()
```

```python
scaled_testX = scaler.transform(test[cols])
```

```python
model.predict(test[cols])
```

```python
result=pd.DataFrame(test['Id'], columns=['Id'])
result['SalePrice']=model.predict(scaled_testX).ravel()
result['SalePrice']=np.round(np.expm1(result['SalePrice']),2)
result
```

```python
result.to_csv('submission.csv', index=False)
```
